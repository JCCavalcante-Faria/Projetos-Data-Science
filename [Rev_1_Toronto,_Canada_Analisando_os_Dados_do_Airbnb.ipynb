### **Q2. Qual a porcentagem de valores ausentes no *dataset*?**
A qualidade de um dataset está diretamente relacionada à quantidade de valores ausentes. É importante entender logo no início se esses valores nulos são significativos comparados ao total de entradas.

* É possível ver que a coluna **neighbourhood_group** possui 100% dos seus valores faltantes.
* As variáveis **reviews_per_month** e **last_review** possuem valores nulos em quase 30% das linhas.
* As variáveis **name** e **host_name** têm menos de 0,1% dos valores nulos.

Um valor **ausente** ou **nulo** pode significar várias coisas diferentes. Talvez o campo não fosse aplicável, o evento não aconteceu ou os dados não estavam disponíveis. Pode ser que a pessoa que inseriu os dados não sabia o valor certo, ou não se preocupou que um campo não foi preenchido.

No entanto, há muitos cenários de análise de dados nos quais os valores ausentes fornecem informações importantes. O significado dos valores ausentes depende em grande parte do contexto. 

```
# ordenar em ordem decrescente as variáveis por seus valores ausentes
(df. isnull().sum()/ df.shape[0]).sort_values (ascending=False)
```

Nesse caso vamos optar pela remoção das colunas neighbourhood_group e license pois tem poucos ou nenhum dado e vamos substituir por zero os dados ausentes nas colunas reviews_per_month e last_review pois são dados que não tem poder de influenciar essa análise, emfim, vamos excluir as linhas de entrada nulas de name e host_name por que são menos de 1% das entradas.

```
# Excluir colunas com dados faltantes
df.drop(columns=['neighbourhood_group', 'license'], inplace=True)
     

# Substituir as entradas nulas por zero
df['reviews_per_month'] = df['reviews_per_month']. fillna (0)
df['last_review'] = df['last_review']. fillna (0)

     

# Excluir as linhas da entrada 'name''host_name' que possuem valores ausentes
df.dropna(subset=['name'], axis=0, inplace=True)
df.dropna(subset=['host_name'], axis=0, inplace=True)

df.info()
```

Agora fica claro que não temos valores ausentes.

```
# Verificar se ainda restam mais valores ausentes
(df.isnull().sum()).sort_values(ascending=False)
```

### **Q3. Qual o tipo de distribuição das variáveis?** 
O que é um histograma?

Um histograma é uma espécie de gráfico de barras que demonstra uma distribuição de frequências. No histograma, a base de cada uma das barras representa uma classe e a altura representa a quantidade ou frequência absoluta com que o valor de cada classe ocorre.



```
# plotar o histograma das variáveis numéricas
df.hist (bins=15, figsize=(15,10));
```

Ao avaliar os histogramas é possível perceber o comportamento anormal na coluna  Price e na coluna minimum_nights onde os valores são elevados devido a dispersão de entradas anormais.

* HÁ OUTLIERS PRESENTES?

Em estatística, outlier, valor aberrante ou valor atípico, é uma observação que apresenta um grande afastamento das demais da série, ou que é inconsistente. A existência de outliers implica, tipicamente, em prejuízos à interpretação dos resultados dos testes estatísticos aplicados às amostras.

Pela distribuição do histograma, é possível verificar indícios da presença de outliers. Olhe por exemplo as variáveis price, minimum_nights e calculated_host_listings_count.

Os valores não seguem uma distruição, e distorcem toda a representação gráfica. Para confirmar, há duas maneiras rápidas que auxiliam a detecção de outliers. São elas:

* Resumo estatístico por meio do método describe()
* Plotar boxplots para a variável.

```
# ver o resumo estatístico das variáveis numéricas
df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].describe()
```

No resumo acima conseguimos obter várias informações, como:
* * A variável `price` possui um preço médio de  179,00 dólares canadenses, um valor mínimo de 0,0 e valor máximo de  51.561,00 dólares canadenses, essa discrepancia fará com que a média seja fora da realidade e é gerada pelos outliers.
* A quantidade mínima de noites (`minimum_nights`) está acima de 365 dias no ano, mais uma consequência dos outliers.



# Identificando limites máximo e mínimo do conjunto minimum_night.
Ao determinar um limite evitamos que dados discrepantes comprometam o resultado final.


```
# identificando os outliers para a variável minimum_nights
q1_minimum_nights = df.minimum_nights.quantile(.25)
q3_minimum_nights = df.minimum_nights.quantile(.75)
IQR_minimum_nights = q3_minimum_nights - q1_minimum_nights
print('IQR da variável minimum_nights: ', IQR_minimum_nights)

# definindo os limites                                       
sup_minimum_nights = q3_minimum_nights + 1.5 * IQR_minimum_nights
inf_minimum_nights = q1_minimum_nights - 1.5 * IQR_minimum_nights

print('Limite superior de minimum_nights: ', sup_minimum_nights)
print('Limite inferior de minimum_nights: ', inf_minimum_nights)
```

* Após a identificação do limites encontramos o limite para minimum_nights de 28 dias, em seguida veremos o comportamento no gráfico.

```
# minimum_nights
fig, ax = plt.subplots(figsize=(15,3))
df.minimum_nights.plot(kind='box', vert=False, figsize=(15, 3))
ax.set_title('Dataset Original - minimum_nights')
plt.show()

# ver quantidade de valores acima de 28 dias para minimum_nights
print("minimum_nights: valores acima de 28:")
print("{} entradas".format(len(df[df.minimum_nights > 28])))
print("{:.4f}%".format((len(df[df.minimum_nights > 28]) / df.shape[0])*100))

```

 # Identificando os limites maximo e mínimo do conjunto Price.
  
  



```
# identificando os outliers para a variável price
q1_price = df.price.quantile(.25)
q3_price = df.price.quantile(.75)
IQR_price = q3_price - q1_price
print('IQR da variável price: ', IQR_price)

# definindo os limites                                       
sup_price = q3_price + 1.5 * IQR_price
inf_price = q1_price - 1.5 * IQR_price

print('Limite superior de price: ', sup_price)
print('Limite inferior de price: ', inf_price)
```

* Aqui após identificar os limites, vemos que o preço maximo práticado é de 385 dólares canadenses, em seguida veremos como se comporta no gráfico.

```
#  boxplot price
fig, ax = plt.subplots(figsize=(15,3))
df.price.plot(kind='box', vert=False, figsize=(15, 3),)
ax.set_title('Dataset Original - Price')
plt.show()

# ver quantidade de valores acima de 385 para price
print("\nprice: valores acima de 385")
print("{} entradas".format(len(df[df.price > 385])))
print("{:.4f}%".format((len(df[df.price > 385]) / df.shape[0])*100))
```

 Identificamos outliers nas variáveis price e minimum_nights, faremos a limpeza dos dados com base em novos parâmetros.

```
# remover os outliers em um novo DataFrame
df_clean = df.copy()
df_clean.drop(df_clean[df_clean.price > 385].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.minimum_nights > 28].index, axis=0, inplace=True)


print('Dados antes da limpeza: ', df.shape)
print('Dados após a limpeza: ',df_clean.shape)
```

#Vamos ver ver como fica um BOXPLOT do conjunto limpo.

```
# verificando conjunto limpo
fig, ax = plt.subplots(figsize=(15,3))
df_clean.minimum_nights.plot(kind='box', vert=False);
ax.set_title('Dataset Limpo - minimum_nights')
plt.show()
print("Tamanho original: {} colunas".format(df.shape[0]))
print("Tamanho atual: {} colunas".format(df.shape[0]))
```

Agora, faremos os histogramas onde será possível ver a distribuição nas variáveis price e minimum_night sem interferência dos dados anormais (outliers), note que, dentro dos padrões o gráfico representa uma distribuição mais próxima da realidade.

### **Q4. Qual a correlação existente entre as variáveis**
Correlação significa que existe uma relação entre duas coisas. No nosso contexto, estamos buscando relação ou semelhança entre duas variáveis.

Essa relação pode ser medida, e é função do coeficiente de correlação estabelecer qual a intensidade dela. Para identificar as correlações existentes entre as variáveis de interesse, vou:

*  Criar uma matriz de correlação
*  Gerar um heatmap a partir dessa matriz, usando a biblioteca seaborn

```
# criar uma matriz de correlação
corr = df_clean[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].corr()

display(corr)



```

```
# plotar um heatmap a partir das correlações
sns.heatmap(corr, cmap='RdBu', fmt='.2f', square=True, linecolor='white', annot=True);
```

### **Q6. Qual a localidade mais cara de Toronto?**
Uma maneira de se verificar uma variável em função da outra é usando groupby(). No caso, queremos comparar os bairros (neighbourhoods) a partir do preço de locação.
